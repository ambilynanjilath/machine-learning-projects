<h1>Exploratory Data Analysis</h1>

histogram

```python

import matplotlib.pyplot as plt
import seaborn as sns
num_cols = len(df1.columns[:-1])
num_rows = num_cols // 2 + num_cols % 2
fig, axes = plt.subplots(nrows=num_rows, ncols=2, figsize=(12, num_rows * 4))
for i, feature in enumerate(df1.columns[:-1]):
    row = i // 2
    col = i % 2
    ax = axes[row][col]
    sns.histplot(data=df1, x=feature, bins=10, kde=True, ax=ax)
    ax.set_xlabel(feature)
    ax.set_ylabel('Frequency')
    ax.set_title('Histogram of ' + feature)
if num_cols % 2 != 0:
    axes[-1, -1].axis('off')
plt.tight_layout()
plt.show()
```

scatter plot

```python
fig, axes = plt.subplots(nrows=num_rows, ncols=2, figsize=(12, num_rows * 4))
plot_num=0
for i in range(df1.shape[1] - 1):
    for j in range(i + 1, df1.shape[1]):
        if plot_num < num_cols:
            ax = axes[plot_num // 2, plot_num % 2]
            sns.scatterplot(data=df1, x=df1.columns[i], y=df1.columns[j], hue='Species', palette='Set1', ax=ax)
            ax.set_xlabel(df1.columns[i])
            ax.set_ylabel(df1.columns[j])
            ax.set_title('Scatter Plot: ' + df1.columns[i] + ' vs ' + df1.columns[j])
            ax.legend(title='Species')
            plot_num += 1
        else:
            break
if num_cols % 2 != 0:
    axes[-1, -1].axis('off')

plt.tight_layout()
plt.show()
```

split the data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_test
```

```python
y_train
```

standardscalar is implemented to normalize the features

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

```python
x_train
```

```python
x_test
```

<h1>Model Creation</h1>

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greys')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
score=accuracy_score(y_test,y_pred)
print("Accuracy Score is",score)
```

```python
print("Classification Report is")
print(classification_report(y_test,y_pred))
```
