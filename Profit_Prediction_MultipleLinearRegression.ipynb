**Seperate x as input feature and y as class label**

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

**Applied onehot encoding to transform the object values**

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_trans.fit_transform(x)
```

```python
x
```

**Split the data into training and testing**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# Model Creation

**Created model using LinearRegression**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

**Create dataframe using Actual value,Predicted value, and the difference**

```python
df1=pd.DataFrame({'actual value':y_test,'predicted_value':y_pred,'difference':y_test-y_pred})
df1
```

**Residual Plot**

```python
residuals=y_test-y_pred
plt.scatter(y_pred,residuals,alpha=0.6)
plt.xlabel("Predicted Value")
plt.ylabel("Actual Value")
plt.title("Residual PLot:Predicted Vs Actual Value")
plt.axhline(y=0,color='r',linestyle='--')
plt.show()
```

**Histogram of Residuals**

```python
plt.figure(figsize=(8,6))
plt.hist(residuals,bins=30,edgecolor='black')
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Histogram of Residuals")
```

**Line Plot of Actual value Vs Predicted value**

```python
plt.plot(range(len(y_test)),y_test,label='Actual Value')
plt.plot(range(len(y_test)),y_pred,label='Predicted Value')
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Actual Value Vs Predicted Value")
plt.legend()
plt.show()
```

**Printing slope and y intercept**

```python
print("slope is")
print(model.coef_)
print("y intercept is")
print(model.intercept_)
```

# Performance Evaluation

**Performance evaluation using Mean Absolute Error,Mean Absolute Percentage Error,Mean Squared Error,Root Mean Squared Error,R2 Score**

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score,mean_squared_error
mae=mean_absolute_error(y_test,y_pred)
mape=mean_absolute_percentage_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
score=r2_score(y_test,y_pred)
print("Mean Absolute Error",mae)
print("Mean Absolute Percentage Error",mape)
print("Mean Squared Error",mse)
print("Root Mean Squared Error",rmse)
print("R2 Score",score)
```
