```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/Mall_Customers.csv')
df
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,3:5]
x
```

```python
from sklearn.cluster import KMeans
wcss=[] #list created to store varience
for i in range(1,11):
  data=KMeans(n_clusters=i,init="k-means++",random_state=42)
  data.fit(x)
  wcss.append(data.inertia_)#inertia used to calculate varience
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("no of clusters")
plt.ylabel("varience")
plt.title("clusters vs varience")
```

```python
#here from 5 it constant.so 5 takes as cluster point
```

```python
data1=KMeans(n_clusters=5,init="k-means++",random_state=42)
y_data=data1.fit_predict(x)
y_data
```
