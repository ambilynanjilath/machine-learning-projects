**Perform get dummies encoding to transform the categorical columns**

```python
df2=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
print(df2)
```

**Concatinating with the old dataframe**

```python
df3=pd.concat([df,df2],axis=1)
print(df3)
```

**Droping unnecessary columns**

```python
a=df3.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price'],axis=1)
print(a)
```

**Unit conversion for numeric columns**

```python
a['Mileage']=a['Mileage'].str.replace('km/kg','')
a['Mileage']=a['Mileage'].str.replace('kmpl','')
a['Engine']=a['Engine'].str.replace('CC','')
a['Power']=a['Power'].str.replace('bhp','')
```

```python
print(a)
```

**Replacing null with 0:Missing value handling**

```python
a['Mileage']=a['Mileage'].str.replace('null','0')
a['Power']=a['Power'].str.replace('null','0')
a['Engine']=a['Engine'].str.replace('null','0')
print(a)
```

```python
a.dtypes
```

**Datatype conversion**

```python
a['Engine']=a['Engine'].astype(float)
a['Power']=a['Power'].astype(float)
a['Mileage']=a['Mileage'].astype(float)
```

```python
a.dtypes
```

```python
a.isna().sum()
```

**Replacing 0 with NaN:Handling missing value**

```python
a.loc[a.Engine==0,'Engine']=np.NaN
a.loc[a.Mileage==0,'Mileage']=np.NaN
a.loc[a.Power==0,'Power']=np.NaN
```

```python
a.isna().sum()
```

**Fill the missing value with mean and mode**

```python
x1=a['Engine'].mean()
a['Engine'].fillna(x1,inplace=True)
x2=a['Power'].mean()
a['Power'].fillna(x2,inplace=True)
x3=a['Mileage'].mean()
a['Mileage'].fillna(x3,inplace=True)
a['Seats']=a['Seats'].fillna(a['Seats'].mode()[0])
```

```python
a.isna().sum()
```

**seperate x_train as input feature**

```python
x_train=a.drop(['Price','Fuel_Type_Electric'],axis=1)
print(x_train)
```

**y_train as class label**

```python
y_train=a['Price']
y_train
```

# Model Creation

**Model  creation using linearregression**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

**Residual Analysis: Scatter Plot of Predicted Values vs. Residuals for Training Data**

The plot illustrates the differences between the predicted values and the actual target values (residuals) for the training data, providing valuable insights into the model's performance without using a separate y_test for evaluation.

```python
from sklearn.model_selection import learning_curve
residuals = y_train - model.predict(x_train)
plt.scatter(model.predict(x_train), residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```
