**Applied LabelEncoding to transform the categorical values**

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Sex']=lb.fit_transform(df['Sex'])
df['BP']=lb.fit_transform(df['BP'])
df['Cholesterol']=lb.fit_transform(df['Cholesterol'])
```

```python
df
```

**Seperate x as input feature and y as class label**

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

**Split the data into training and testing**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

**Applied MinMaxScalar for data Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Model Creation

**Created model using DecisionTreeClassifier**

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy',splitter='best')
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

**Decision Tree Visualization**

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(10,8))
plot_tree(model, filled=True, feature_names=x.columns, class_names=y.unique())
plt.title('Decision Tree Visualization for Drug Classification')
plt.show()
```

**Feature Importance Bar Plot**

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
feature_importance = model.feature_importances_
sns.barplot(x=feature_importance, y=x.columns)
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Decision Tree Feature Importance')
plt.show()
```

# Performance Evaluation

**Confusion Matrix Visualization**

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='YlGnBu', xticklabels=y.unique(), yticklabels=y.unique())
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
print("Accuracy Score",accuracy_score(y_test,y_pred))
```

```python
print("Classification Report")
print(classification_report(y_test,y_pred))
```
